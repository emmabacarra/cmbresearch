<u><b>About the Spectrum</b></u>

WMAP data includes the temperature and polarization measurements of the CMB

CMB Temperature Variations:
- WMAP measures the temperature of CMB radiation across the sky.
- It shows slight variations or fluctuations, not uniform (anisotropic)
- These fluctuations resemble ripples on a pond.

Causes of Temperature Variations:
- Variations in gas density or velocity at the surface of last scatter.
- Variations in the gravitational potential at the surface of last scatter and along photon path.
- Different physical phenomena cause each effect.

Angular Fluctuation Spectrum:
- CMB temperature measured in smaller patches of the sky.
- Represented by an "angular fluctuation spectrum".

Variations in Gravity (Large Scales):
- Main fluctuations due to gravity strength variations at last scatter.
- Dark matter density largely determines gravitational potential variations.
- Early universe likely featureless, inflation may have caused initial variations.

Imprint of Sound Waves (Small Scales):
- Sound waves in ionized hydrogen gas leave imprints.
- Regions of compression in sound waves appear brighter or hotter.
- Decoupling from hydrogen gas freezes out waves, leaving imprints.

<u><b>About the Data Calibration</b></u>

- COBE Sky Map: uncorrected dipole temperature shift across the sky.
- Radiometric Gain: Proportionality constant between input temperature difference and output voltage.

WMAP Microwave System:
- Consists of ten 4-channel 'differencing assemblies'.
- Each assembly receives signals from a pair of feeds observing the sky.
- Output is proportional to temperature (brightness) difference between the two lines of sight.

Calibration Process:
- Intrinsic random asymmetries in assemblies generate offsets in output data.
- Signals calibrated against known microwave sources for accuracy.

Dipole Effect:
- Result of Earth, solar system, and galaxy movement through the universe.
- Similar to change in sound pitch passing by a source in a car or train.
- CMB temperature 6.706 mK brighter in one sky direction than the opposite.

Data Calibration:
- Raw output data compared to expected signal from known dipole anisotropy.
- Calibrates data to true temperature differences on the sky.
