```python
import astropy
from astropy.io import fits
from astropy.visualization import ZScaleInterval, SqrtStretch, ImageNormalize, HistEqStretch
import numpy as np

# from astropy.cosmology import WMAP9 as wmap
# from astropy.cosmology import Planck18 as planck
import pandas as pd

import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import display, HTML, Image
```

#### Ka1 For the First Year of Data

```python
Image('my first pic of cmb!.png', width=400)
```

```python
# opening the fits file
with fits.open('wmap_data/wmap_mollweide_imap_r9_yr1_Ka1_v5.fits') as hdul:
    print(hdul[0].header['COMMENT'])
    datak1a = hdul[1].data

# Convert data to float type and replace NaN values
datak1a = np.nan_to_num(datak1a.astype(float))

# normalizing the image
normk1a = ImageNormalize(datak1a, interval=ZScaleInterval(), stretch=SqrtStretch())

# plotting the image
plt.imshow(datak1a, cmap='coolwarm', norm=normk1a)
plt.show()
    
```

#### K1 For the Ninth Year of Data

```python
Image('second pic of cmb.png', width=400)
```

```python
with fits.open('wmap_data/wmap_mollweide_imap_r9_yr9_K1_v5.fits') as hdul:
    datak1 = hdul[1].data

datak1 = np.nan_to_num(datak1.astype(float))
vmin, vmax = np.percentile(datak1, [40, 83])
normk1 = ImageNormalize(vmin=vmin, vmax=vmax, stretch=HistEqStretch(datak1))
plt.imshow(datak1, cmap='Spectral_r', norm=normk1)
plt.show()
```
